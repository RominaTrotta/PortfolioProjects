```python
from wordcloud import WordCloud

plt.figure(figsize=(12,11))
wc = WordCloud(min_font_size = 10,
               max_words=100,
               height = 1000, width=2000,
               colormap="Set2", background_color='white',
               random_state = 1,
               collocations=False).generate(text)

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('simple_wc.jpg', dpi=300, pad_inches=0, bbox_inches='tight')
plt.show()
```

```python
#  Using a specific random color of our choice

def one_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None):
    """function that returns the hsl: Hue-Saturation-Lightness.
    Choose from: https://hslpicker.com/
    
    Range of this params:
    h =  0 - 360
    s =  0 - 100
    l =  0 - 100"""  
    
    h=360
    s=84
    l=31
    
    return "hsl({}, {}%, {}%)".format(h, s, l)

mask = np.array(Image.open(r'D:\Romi\Python_projects_scripts\Olympics\oly.png'))
mask_colors = ImageColorGenerator(mask)

wc = WordCloud(font_path=font_path,
               mask=mask, 
               width=mask.shape[1],
               height=mask.shape[0], 
               max_words=100, 
               min_font_size=10,
               random_state=1, 
               background_color="white", 
               color_func=one_color_func,# instead of mask_colors
               collocations=False).generate(text)

plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
#  Using many specific random colors of our choice

def multi_color_func(word=None, font_size=None,
                     position=None, orientation=None,
                     font_path=None, random_state=None):
    """function that returns the hsl: Hue-Saturation-Lightness.
    Choose from: https://hslpicker.com/
    
    Range of this params:
    h =  0 - 360
    s =  0 - 100
    l =  0 - 100"""  
    
    h = 40 # 0 - 360
    s = 100 # 0 - 100
    l = 50 # 0 - 100
    colors = [[14, 77, 50],
              [25, 74, 85],
              [260, 43, 84],
              [120, 11, 79]]
    rand = random_state.randint(0, len(colors) - 1)

    return "hsl({}, {}%, {}%)".format(colors[rand][0], colors[rand][1], colors[rand][2])

mask = np.array(Image.open(r'D:\Romi\Python_projects_scripts\Olympics\helmet.png'))

wc = WordCloud(font_path=font_path,
               mask=mask, 
               width=mask.shape[1],
               height=mask.shape[0], 
               max_words=100, 
               min_font_size=10,
               random_state=1, 
               background_color="white", 
               color_func=multi_color_func,             
               collocations=False).generate(text1)

plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
